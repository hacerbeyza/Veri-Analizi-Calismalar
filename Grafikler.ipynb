```python
import pandas as pd
import numpy as np
veri5=pd.read_excel("data5.xlsx")
veri5.head()
```

```python
import matplotlib.pyplot as plt

renkler=["Green", "Blue", "Teal"]
plt.bar(veri5.City, veri5.cogs, color=renkler)
plt.title("Market Satışları", fontsize=14)
plt.xlabel("City", fontsize=14)
plt.ylabel("Cogs")
plt.rcParams["figure.figsize"]=(5,5)
plt.grid(True)
plt.show()
```
